### 2.1. X et Y
</div>

```python
X, y = mnist["data"], mnist["target"]
```

```python
y = y.astype(np.uint8)
```

```python
from sklearn.manifold import TSNE
from mlinsights.mlmodel import PredictableTSNE
```

```python
random_state=0
verbose=1
```

```python
X_tsne = TSNE(random_state = random_state, n_components=2,verbose=verbose, perplexity=40, n_iter=300, init='random', learning_rate=200.0).fit_transform(X)
```

```python
X_train_tsne, X_test_tsne, Y_train, Y_test = X_tsne[:60000], X_tsne[60000:], y[:60000], y[60000:]
```
